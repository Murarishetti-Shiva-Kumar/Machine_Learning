## Matplotlib Example workflow

```python
#0. import required libraries
%matplotlib inline
import matplotlib.pyplot as plt

#1. Prepare data
x=[1,2,3,4]
y=[11,22,33,44]

#2. plot
fig, ax=plt.subplots(figsize=(10,10))#width,height

#3. plot data to axis
ax.plot(x,y)

#4. Customize plot
ax.set(title="sample matplot lib",
      xlabel="x-axis",
      ylabel="y-axis")

#5. Save & print
fig.savefig("sample_matplotlib.png")
```

#when to use pyplot method vs matplotlib OO Method

#pyplot is for something quick
#OO Method is used something more advanced

```python
age_over_50=heart_disease["age"]>50
```

```python
age_over_50
```

```python
len(age_over_50)
```

```python
age_over_50=heart_disease[heart_disease["age"]>50]
```

```python
age_over_50
```

```python
len(age_over_50)
```

```python
age_over_50.plot(x="age",
                y="chol",
                kind="scatter")
```

```python
#pyplot method
age_over_50.plot(x="age",
                y="chol",
                kind="scatter",
                c="target")
```

```python
#using style colors with OO Method

#Setting the style
plt.style.use("seaborn-whitegrid")

fig,(ax0,ax1)=plt.subplots(nrows=2,
                          ncols=1,
                          figsize=(10,6),
                          sharex=True)

#Addind data to ax0 graph
scatt1=ax0.scatter(x=age_over_50["age"],
                   y=age_over_50["chol"],
                   c=age_over_50["target"],
                  cmap="winter")

#Customize the graph
ax0.set(title="Heart Diseases & cholestrol Levels",
#        xlabel="Age",
       ylabel="Cholestrol")

ax0.set_xlim(['50','80'])

#Adding legends
ax0.legend(*scatt1.legend_elements(),title="Target");

#Adding a mean line
ax0.axhline(y=age_over_50["chol"].mean(),
           linestyle="--")

#Creating ax1 graph for thalach
scatt2=ax1.scatter(x=age_over_50["age"],
                  y=age_over_50["thalach"],
                  c=age_over_50["target"],
                  cmap="winter")

#Customize the graph
ax1.set(title="Heart Diseases & Max Heart Rate",
       xlabel="Age",
       ylabel="Thalach")

#Adding the legend
ax1.legend(*scatt2.legend_elements(),title="Target")

#Adding the horizontal line
ax1.axhline(y=age_over_50["thalach"].mean(),
           linestyle="--")

#Adding Title to the figure
fig.suptitle("Heart Disease Analysis",
            fontsize="12",
            fontweight="bold");
```
